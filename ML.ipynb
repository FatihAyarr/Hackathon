# Kullancağımız kütüphaneleri import ediyoruz.

```python
import numpy as np
import pandas as pd
import sklearn
```

# Verisetimizi okuyoruz.

```python
df = pd.read_csv('Placement_Data_Full_Class.csv')
df.head()
```

# Modelimizde kullanmayacağımız kolonları datasetimizden düşürüyoruz.

```python
df.drop(['sl_no','ssc_b','hsc_b','etest_p','salary'], axis=1, inplace=True)
df.head()
```

# Kategorik ve numerik kolonları listeleyip kategorik, numerik ve target dataframeleri oluşturuyoruz.

```python
col_num = ['ssc_p','hsc_p','degree_p','mba_p'] #liste

col_cat = ['gender','hsc_s','degree_t','specialisation'] #liste

col_tar = 'status' #string

df_num = df[col_num]
df_cat = df[col_cat]
y = df[col_tar]
```

# Kategorik verilere one-hot encoding uyguluyoruz.# 

```python
df2 = df_cat
df2.head()
```

```python
sg_dum = pd.get_dummies(df2)
sg_dum.head()
```

# One-hot encoding uyguladığımız verilerimiz ile numerik verilerimizi birleştiriyoruz.

```python
X = pd.concat([df_num,sg_dum],axis=1)
print(X.isna().sum().sum()) # 0 olmalıdır.
X.head()
```

# Numerik verilerimize normalizasyon uyguluyoruz.

```python
from sklearn.preprocessing import Normalizer
datafr = Normalizer().fit_transform(X)
datafr
```

```python
X = pd.DataFrame(datafr,columns=['ssc_p', 'hsc_p', 'degree_p', 'mba_p', 'gender_F', 'gender_M',
       'hsc_s_Arts', 'hsc_s_Commerce', 'hsc_s_Science', 'degree_t_Comm&Mgmt',
       'degree_t_Others', 'degree_t_Sci&Tech', 'specialisation_Mkt&Fin',
       'specialisation_Mkt&HR'])
X.head()
```

# Gerekli kütüphanelerimizi import edip train ve test datalarını oluşturuyoruz.

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.35,random_state=10)
```

# Farklı ML algoritmaları ile modelimizi eğitip doğruluk oranlarına bakıyoruz.

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)
accuracy*100
```

```python
from sklearn.naive_bayes import GaussianNB

clf2 = GaussianNB()
clf2.fit(X_train,y_train)

y_pred = clf2.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)
accuracy*100
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf3 = KNeighborsClassifier(n_neighbors=7)
clf3.fit(X_train,y_train)

y_pred = clf3.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)
accuracy*100
```

```python
from sklearn.svm import SVC

clf4 = SVC(kernel='rbf',probability=True)
clf4.fit(X_train,y_train)

y_pred = clf4.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)
accuracy*100
```

```python
from sklearn.ensemble import RandomForestClassifier

clf5 = RandomForestClassifier(max_depth=10)
clf5.fit(X_train,y_train)

y_pred = clf5.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)
accuracy*100
```
